5. What is mutability in Python?
- Mutability in Python refers to the ability of an object to change its state or contents after it has been created. In other words, mutable objects can be modified, whereas immutable objects cannot be changed once they are created.
- Example: Lists, Dictionaries.

6. Why are lists mutable, but tuples are immutable?
- The difference in mutability between lists and tuples in Python is intentional and rooted in their design and usage principles.
- Lists
  - Lists are designed to be flexible and dynamic. You can modify them by adding, removing, or changing elements. This makes lists suitable for collections of items that need to be frequently updated.
  - Think of lists as grocery carts—you can keep adding or removing items based on what you need.
- Tuples
  - Tuples, on the other hand, are immutable. Once you create a tuple, you cannot change its content. This immutability makes tuples reliable and predictable because their contents remain constant.
  - Immutable objects can be optimized by the Python interpreter in ways that mutable objects cannot. This can lead to performance improvements in certain cases.
  - Consider tuples as the packed groceries you bring home. The contents are fixed and ready to be used as-is.

8. What are logical operators in Python?

  - **and Operator**

  - **or Operator**

9. What is type casting in Python?
- Type casting in Python, also known as type conversion, is the process of converting one data type to another. This can be done explicitly using built-in functions or implicitly by the Python interpreter under certain circumstances.

10. What is the difference between implicit and explicit type casting?
- Implicit Type Casting
  - The Python interpreter automatically converts one data type to another without explicit instructions from the programmer.
  - This often happens in operations where Python needs to ensure compatibility between different data types.
  - Generally, implicit type casting happens in a way that avoids loss of data or precision.
- Explicit Type Casting
  - The programmer explicitly converts one data type to another using built-in functions.
  - Provides more control over the conversion process, ensuring that it happens exactly when and where intended.
  - There is a risk of losing data or precision if the conversion is not handled properly.

12. How does the elif statement work?
- The elif statement in Python stands for "else if" and is used to check multiple conditions in sequence. It's a way to handle more than two possible outcomes in a conditional structure, without having to nest multiple if and else statements.
- Here’s how the elif statement works:
  - First Condition (if):
    - The if statement is evaluated first.
    - If the if condition is True, its corresponding block of code is executed, and the rest of the elif and else statements are skipped.
  - Additional Conditions (elif):
    - If the if condition is False, the program checks the next condition specified in the elif statement.
    - If the elif condition is True, its corresponding block of code is executed, and the remaining elif and else statements are skipped.
  - Final Condition (else):
    - If none of the if or elif conditions are True, the else block is executed, if it exists.

13. What is the difference between for and while loops?
- Both for and while loops in Python are used for repeating a block of code multiple times, but they differ in their structure and usage.
- for Loop
  - Iterates Over a Sequence: A for loop is primarily used to iterate over a sequence (like a list, tuple, string, or range).
  - Fixed Iteration: The number of iterations is determined by the length of the sequence or the specified range.
- while Loop
  - Based on a Condition: A while loop continues to execute as long as a specified condition remains True.
  - Potentially Infinite: The number of iterations is not fixed and depends on the condition. The loop can run indefinitely if the condition never becomes False.

In this case, the while loop is more suitable because you don't know how many attempts the user will need to enter the correct password.

9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
user_input = input("Enter a string: ")
try:
    int_value = int(user_input)
    print(f"String to Integer: {int_value}")
except ValueError:
    print("Cannot convert to Integer")
try:
    float_value = float(user_input)
    print(f"String to Float: {float_value}")
except ValueError:
    print("Cannot convert to Float")
bool_value = user_input.lower() in ['true', '1', 'yes']
print(f"String to Boolean: {bool_value}")
```

10. Write code to demonstrate type casting with list elements.

```python
original_list = ["1", "2", "3.5", "True", "False"]
int_list = []
float_list = []
bool_list = []
for item in original_list:
    try:
        int_list.append(int(item))
    except ValueError:
        int_list.append(None)
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(None)
    bool_list.append(item.lower() in ['true', '1', 'yes'])
print("Original list:", original_list)
print("Converted to integers:", int_list)
print("Converted to floats:", float_list)
print("Converted to booleans:", bool_list)
```

14. Write a program to reverse a string using a while loop.

```python
user_input = input("Enter a string: ")
reversed_string = ""
index = len(user_input) - 1
while index >= 0:
    reversed_string += user_input[index]
    index -= 1
print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
number = int(input("Enter a number: "))
factorial = 1
i = number
while i > 0:
    factorial *= i
    i -= 1
print(f"The factorial of {number} is: {factorial}")
```
