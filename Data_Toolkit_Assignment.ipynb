6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays and Python lists are both used to store and manipulate collections of elements, but they have some key differences that are listed below.
  - Performance: Numpy Array is highly Optimized for performance and can handle large datasets efficiently as compared to Python lists.
  - Data Type: NumPy Arrays have all elements must be of the same data type (e.g., all integers, all floats) where as Python Lists Can store elements of different data types (e.g., integers, floats, strings, etc.).
  - Memory Consumption: Numpy Array Consume less memory as they use a fixed data type and are stored in contiguous memory blocks as compared to Python lists.
  - Functionality: NumPy Arrays Provide a wide range of mathematical and statistical functions, making them suitable for numerical and scientific computing where as Python Lists Offer basic list operations like appending, slicing, and indexing, but lack built-in functions for advanced mathematical operations.

8. What does the term “vectorized operation” mean in NumPy?
- In NumPy, a vectorized operation refers to the process of performing element-wise operations on entire arrays (vectors) without the need for explicit loops. Vectorized operations are optimized for performance and take advantage of low-level optimizations provided by NumPy and the underlying hardware, resulting in faster and more efficient computations.

```python
#Example:
#Without Vectorization
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
result = np.zeros_like(arr)
for i in range(len(arr)):
    result[i] = arr[i] + 10
print(result)
```

```python
#With Vectorization
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
result = arr + 10
print(result)
```

17. Explain the difference between apply() and map() in Pandas.
- apply():
  - Can be used on DataFrames and Series.
  - Applies a function along a specified axis (rows or columns) of a DataFrame or to each element of a Series.
  - More versatile as it can apply functions to entire DataFrame rows or columns.
- map():
  - Can be used on Series only.
  - Applies a function element-wise to each value in the Series.
  - Best suited for transforming or mapping individual values.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.figure()
sns.heatmap(corr_matrix, annot = True,cmap = 'coolwarm')
plt.show()
```
