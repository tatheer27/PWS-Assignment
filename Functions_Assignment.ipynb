6. Explain the concept of generators in Python and how they are defined.
- As we know in regular function, the function calculates and return in one go, since it will return in one go, memory usuage will be more and it will take a lot of time if there is lot of computation as all the computation will be done first and then the calculation will be returned in one go.
- So we need generator to generate result one by one instead of one go.
- General function uses return statement but a generator function uses **yield** statement.

```python
#example
def sq_num(n):
  for i in range(n):
    yield i**2
s = sq_num(100)
print(next(s))
print(next(s))
print(next(s))
print(next(s))
```

7. What are the advantages of using generators over regular functions?
- Generators provide memory efficiency, lazy evaluation, improved performance, and simplified code for iterating over large datasets or streams of data.

```python
#Generator function
def count(n):
    c = 1
    while c <= n:
        yield c
        c += 1

counter = count(5)
print(next(counter))
print(next(counter))
```

```python
#Regular Function
def count(n):
    return list(range(1, n + 1))

counter = count(5)
print(counter)
```

9. Explain the purpose and usage of the ‘map()” function in Python.
- The map() function in Python is a built-in function that allows you to apply a given function to each item of an iterable (like a list) and returns a map object (which can be converted into a list, tuple, etc.).
- Purpose:
  - The main purpose of map() is to transform one iterable into another by applying a function to each of its elements.
- Usage:

```python
#1. Single iterable
n=[1,2,3]
s=map(lambda x:x**2,n)
print(list(s))
```

```python
#2. Multiple iterables
n1=[2,3,4]
n2=[5,6,7]
s=map(lambda x,y:x*y,n1,n2)
print(list(s))
```

10. What is the difference between ‘map()", ‘reduce()’, and “filter()" functions in Python?
- **Map()**
  - Executes a specified function for each of item of an iterable.
- **Reduce()**
  - It is used for folding/reduction.
  - Reduce will always take two arguments.
- **Filter()**
  - It is used to filter elements from an iterable based on some condition.

```python
#map()
n1=[2,3,4]
n2=[5,6,7]
s=map(lambda x,y:x*y,n1,n2)
print(list(s))
```

```python
#Reduce()
from functools import reduce
n=[1,2,3,8,5,6]
s=reduce(lambda a,b:a if a>b else b,n)
print(s)
```

```python
#Filters
from functools import reduce
n=[1,2,3,4,5,6,7,8]
s=filter(lambda x:x%2==0,n)
print(list(s))
```

```python
def primecheck(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True

for i in range(1,201):
  if primecheck(i):
    print(f"{i} is Prime Number")
  else:
    print(f"{i} is not a Prime Number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.n1, self.n2 = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.n1
        elif self.count == 1:
            self.count += 1
            return self.n2
        else:
            self.n1, self.n2 = self.n2, self.n1 + self.n2
            self.count += 1
            return self.n2

fib_iterator = FibonacciIterator(10)
for term in fib_iterator:
    print(term)
```

10. Create a Python program that uses “filter()” to remove all the vowels from a given string.

```python
s="Anuj Shandilya"
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, s))
print(remove_vowels(s))
```
