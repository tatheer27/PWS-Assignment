7. How tuples ensure data integrity in Python?

  - Once a tuple is created, its contents cannot be altered. This means that the data remains constant throughout the program, preventing accidental modifications.

8. What is a hash table, and how does it relate to dictonaries in Python?
- A hash table is a data structure that efficiently stores and retrieves data using key-value pairs. It uses a hash function to convert keys into unique hash values (usually integers) that determine the index where the corresponding value is stored. This allows for quick access to the data.
- Relation to Dictionaries in Python:
  - Python dictionaries are implemented using hash tables.

  - Keys are hashed to generate hash values, which determine where the corresponding values are stored in the dictionary.

  - This allows for efficient lookups, insertions, and deletions.

14. How does the "in" keyword work differently for lists and dictonaries?
- The in keyword works differently for lists and dictionaries in Python.
- Lists
  - Checks if a value is present in the list.
  - Iterates through the list to see if the specified value exists.
  - Slower for large lists due to linear search (O(n)).
- Dictonaries
  - Checks if a key is present in the dictionary.
  - Uses the hash table to quickly determine if the key exists.
  - Faster due to hash table lookup (O(1) average case).

16. What is a nested dictionary, and give an example of its use case?
- A nested dictionary in Python is a dictionary where the values are also dictionaries. This allows for a hierarchical, multi-level structure of data, providing a way to organize and access complex information efficiently.
- Example is coded below

```python
pwstudents = {
    'student1': {
        'name': 'Ajay',
        'age': 20,
        'grades': {
            'math': 90,
            'science': 85
        }
    },
    'student2': {
        'name': 'Anuj',
        'age': 22,
        'grades': {
            'math': 75,
            'science': 80
        }
    }
}

print(pwstudents['student1']['name'])
print(pwstudents['student1']['grades']['math'])
```

17. Describe the time complexity of accessing elements in a dictionary.
- The time complexity of accessing elements in a dictionary is generally O(1) (constant time) on average. This means that the time required to retrieve a value using its key is independent of the size of the dictionary. This efficiency is due to the underlying hash table implementation that dictionaries use.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
- Dictionaries were historically considered unordered because they didn't maintain any specific order for their key-value pairs. Instead, they relied on hash tables for efficient data storage and retrieval, meaning the order of items could be arbitrary and change as the dictionary was modified.
- It Affects Data Retrieval:
  - When iterating over a dictionary, you couldn't rely on items being in the order they were added.
  - The primary advantage was efficiency. Accessing, adding, and removing items in a dictionary was fast (average O(1) time complexity), regardless of the order of items.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- Lists
  - Elements are accessed by their index (position) in the list.
  - Retrieving an element by index is O(1) (constant time). However, searching for an element (if its position is unknown) requires O(n) (linear time) since it involves iterating through the list.
- Dictionaries
  - Elements are accessed by their unique keys.
  - Retrieving a value by its key is O(1) (constant time) on average, thanks to the hash table implementation. This makes lookups very fast.
